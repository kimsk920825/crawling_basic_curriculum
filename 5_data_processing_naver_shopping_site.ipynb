```python
import requests 
from bs4 import BeautifulSoup 
res = requests.get('https://search.shopping.naver.com/search/all?query=%EA%B0%84%EC%9E%A5&cat_id=&frm=NVSHATC',headers={'User-Agent':'Mozilla/5.0'}) 
soup = BeautifulSoup(res.content,'html.parser' )
soy_sauce_lists = soup.select('ul.list_basis')
#print("https://search.shopping.naver.com/search/all?query=%EA%B0%84%EC%9E%A5&cat_id=&frm=NVSHATC")
for soy_sauce_list in soy_sauce_lists:
     final = soy_sauce_list.find('div').find_all('a')
for single in final:
    print(single.get_text())
```
